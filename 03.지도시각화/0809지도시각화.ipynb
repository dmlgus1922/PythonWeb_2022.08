```python
# Open street map에 광주광역시청과 구청을 표시
# 지도 스타일 변화. tiles
# 마커 아이콘 변경. folium icon을 검색하면 다양한 정류를 찾을 수 있다.
# 마커 말고도 Circle, CircleMarker 등 많음
from folium.features import CustomIcon
from folium.features import DivIcon

map = folium.Map(location=[35.159841, 126.851784],tiles="Stamen Toner", zoom_start=13)

folium.Marker(
    location=[35.159841, 126.851784],
    popup=folium.Popup('광주광역시 서구 내방로 111', max_width = 200),
    tooltip='광주광역시청',
    icon = CustomIcon('다운로드.jpg', icon_size = (80, 80)) # 이미지 파일 로드
).add_to(map)

folium.Marker(
    location=[35.14610731846924, 126.92331464592247],
    popup=folium.Popup('광주광역시 동구 서남로 1', max_width = 200),
    tooltip='광주광역시 동구청',
    icon = DivIcon(icon_size=(100,20), icon_anchor = (0,0), 
                   html='<div style="font-size: 14pt; background-color: powderblue"><bold>광주 동구청</bold></div>')
).add_to(map)

folium.Circle(
    radius=200,
    location=[35.152089096162236, 126.88985826388209],
    popup=folium.Popup('광주광역시 서구 농성동 경열로 33', max_width = 200),
    tooltip='광주광역시 서구청',
    color="magenta",
    fill=True,
).add_to(map)

folium.CircleMarker(
    location=[35.13290098068543, 126.90257332634087],
    radius=20,
    popup=folium.Popup('광주광역시 남구 봉선로 1', max_width = 200),
    color="#3186cc",
    tooltip='광주광역시 남구청',
    fill=True,
    fill_color="#3186cc",
).add_to(map)

folium.Marker(
    location=[35.17410046937537, 126.91215706204946],
    popup=folium.Popup('광주광역시 북구 우치로 77', max_width = 200),
    tooltip='광주광역시 북구청'
).add_to(map)

folium.Marker(
    location=[35.13955198869462, 126.79380680502933],
    popup=folium.Popup('광주광역시 광산구 광산로29번길 15', max_width = 200),
    tooltip='광주광역시 광산구청',
    icon = folium.Icon('blue', icon = 'plane')
).add_to(map)

#지도에 제목 넣기
title_html = '<h3 align = "center">광주광역시 주요 행정기관</h3>'
map.get_root().html.add_child(folium.Element(title_html))
map
```
